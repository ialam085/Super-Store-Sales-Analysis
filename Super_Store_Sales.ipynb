# ◻️ Using **Scatterplot**
>- ### Sales vs Profit by *Categories* and *Sub-Categories*

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', hue='Category', style='Sub-Category', data=df)
plt.title('Sales vs Profit Across Different Categories and Sub-Categories', color='r')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend(title='Category & Sub-Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# ◻️ Using **Barplot** and **Subplot**
>- ### Comparison between **Sales by Segment** and **Profit by Segment**

```python
# Grouping data by Segment and aggregating Sales and Profit
segment_summary = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

# Plotting bar charts
plt.figure(figsize=(10, 6))

# Bar chart for Sales
plt.subplot(1, 2, 1)
sns.barplot(x='Segment', y='Sales', hue = 'Segment', data=segment_summary, palette='muted')
plt.title('Sales by Segment', color='r')
plt.ylabel('Total Sales (in millions)', color='r')

# Bar chart for Profit
plt.subplot(1, 2, 2)
sns.barplot(x='Segment', y='Profit', hue = 'Segment', data=segment_summary, palette='viridis')
plt.title('Profit by Segment', color='g')
plt.ylabel('Total Profit', color='r')

plt.tight_layout()
plt.show()
```

# ◻️ Using **Pie** and **Subplot**
>- ### Comparison between **Sales % by Segment** and **Profit % by Segment**

```python
# Plotting pie charts
plt.figure(figsize=(10, 6))

# Pie chart for Sales
plt.subplot(1, 2, 1)
plt.pie(segment_summary['Sales'], labels=segment_summary['Segment'], autopct='%1.2f%%', colors=sns.color_palette('coolwarm', len(segment_summary)))
plt.title('Sales Distribution by Segment',color='g')

# Pie chart for Profit
plt.subplot(1, 2, 2)
plt.pie(segment_summary['Profit'], labels=segment_summary['Segment'], autopct='%1.1f%%', colors=sns.color_palette('coolwarm', len(segment_summary)))
plt.title('Profit Distribution by Segment',color='r')

plt.tight_layout()
plt.show()
```

# ◻️ Using **Barplot**
>- ### Average Profit by **Shipping Mode**

```python
# Grouping data by Ship Mode and calculating average profit
ship_mode_summary = df.groupby('Ship Mode')['Profit'].mean().reset_index()

# Plotting the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Ship Mode', y='Profit', hue = 'Ship Mode', data=ship_mode_summary, palette='muted')
plt.title('Average Profit by Shipping Mode', color='g')
plt.ylabel('Average Profit', color='r')
plt.xlabel('Shipping Mode', color='b')
plt.show()
```

# ◻️ Using **Barplot**
>- ### Total Sales by **Region**

```python
# Grouping data by Region and calculating total sales
region_sales_summary = df.groupby('Region')['Sales'].sum().reset_index()

# Plotting the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Sales', data=region_sales_summary, hue='Region', palette='coolwarm')
plt.title('Total Sales by Region', color='r')
plt.ylabel('Total Sales', color='r')
plt.xlabel('Region', color='r')
plt.show()
```

# ◻️ Using **Barplot** and **Subplot**
>- ### Comparison between **Total Sales by Category** and **Total Profit by Category**

```python
# Grouping data by Category and calculating total sales and total profit
category_sales_summary = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()

# Plotting bar charts
plt.figure(figsize=(10, 5))

# bar chart for Sale
plt.subplot(1, 2, 1)
sns.barplot(x='Category', y='Sales', data=category_sales_summary, hue='Category', palette='muted')
plt.title('Total Sales by Category', color='r')
plt.ylabel('Total Sales', color='r')

# bar chart for Profit
plt.subplot(1, 2, 2)
sns.barplot(x='Category', y='Profit', data=category_sales_summary, hue='Category', palette='muted')
plt.title('Total Profit by Category', color='r')
plt.ylabel('Total Profit', color='r')

plt.tight_layout()
plt.show()
```

# ◻️ Using **Pie** and **Subplot**
>- ### Comparison between **Sales % by Category** and **Profit % by Category**

```python
# Plotting pie charts
plt.figure(figsize=(10, 6))

# Pie chart for Sales
plt.subplot(1, 2, 1)
plt.pie(category_sales_summary['Sales'], labels=category_sales_summary['Category'], autopct='%1.2f%%', colors=sns.color_palette('muted', len(category_sales_summary)))
plt.title('Sales Distribution by Category', color='r')

# Pie chart for Profit
plt.subplot(1, 2, 2)
plt.pie(category_sales_summary['Profit'], labels=category_sales_summary['Category'], autopct='%1.1f%%', colors=sns.color_palette('muted', len(category_sales_summary)))
plt.title('Profit Distribution by Category', color='r')

plt.tight_layout()
plt.show()
```

# ◻️ Using **Barplot**
>- ### **Top 10 Cities by Sales** and **Top 10 Cities by Profit**

```python
# Grouping data by City and calculating total sales and profits
city_sales_profit_summary = df.groupby('City')[['Sales', 'Profit']].sum().reset_index()

# Sorting the data by sales to show the top cities
top_cities_sales = city_sales_profit_summary.sort_values(by='Sales', ascending=False).head(10)

# Sorting the data by profit to show the top cities
top_cities_profit = city_sales_profit_summary.sort_values(by='Profit', ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x='Sales', y='City', data=top_cities_sales, hue='City', palette='viridis')
plt.title('Top Cities by Sales', color='r')
plt.xlabel('Total Sales')
plt.ylabel('City')
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x='Profit', y='City', data=top_cities_profit, hue='City', palette='muted')
plt.title('Top Cities by Profit', color='r')
plt.xlabel('Total Profit')
plt.ylabel('City')
plt.show()
```

# ◻️ Using **Barplot**
>- ### **State-wise Sales Performance** and **State-wise Profit Performance**

```python
# Grouping data by 'State' and calculating the sum of 'Sales' and 'Profit'
state_sales_profit_summary = df.groupby('State')[['Sales', 'Profit']].sum().reset_index()

# Sorting the data to show the states with the highest sales
state_sales_summary = state_sales_profit_summary.sort_values(by='Sales', ascending=False)

# Sorting the data to show the states with the highest profit
state_profit_summary = state_sales_profit_summary.sort_values(by='Profit', ascending=False)

# Plotting the bar chart for Sales
plt.figure(figsize=(12, 8))
sns.barplot(x='Sales', y='State', data=state_sales_summary, hue='State', palette='Reds_d')
plt.title('State-wise Sales Performance', color='r')
plt.xlabel('Total Sales', color='r')
plt.ylabel('State', color='r')
plt.show()

# Plotting the bar chart for Profit
plt.figure(figsize=(12, 8))
sns.barplot(x='Profit', y='State', data=state_profit_summary, hue='State', palette='viridis')
plt.title('State-wise Profit Performance', color='r')
plt.xlabel('Total Profit', color='r')
plt.ylabel('State', color='r')
plt.show()
```

# ◻️ Using **stacked bar chart**
>- ### Contribution of *Sub-Categories* to **Total Sales** and **Profit**

```python
# Group by 'Sub-Category' and sum 'Sales' and 'Profit'
sub_category_sales_profit = df.groupby('Sub-Category')[['Sales', 'Profit']].sum().reset_index()

# Plotting a stacked bar chart
sub_category_sales_profit.set_index('Sub-Category').plot(kind='bar', stacked=True)
plt.title('Contribution of Sub-Categories to Total Sales and Profit', color='r')
plt.xlabel('Sub-Category', color='r')
plt.ylabel('Amount ($)', color='r')
plt.xticks(rotation=45, color='g')
plt.show()
```
